```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
```

1. Sample of values

```python
def  Pos(P,n):
    return poisson.rvs(P, size =n)
```

```python
P =4
n= 500
s = Pos(P,n)
```

2. Computing the Emperical Mean


```python
Em = sum(s)/n
print("Exact mean =",P )
print("Emperical mean =" ,Em)
```

3.Probability

```python
px = []

for i in s:
    px+= [np.count_nonzero(s==i)/n]
px= np.array(px)    
```

4. Compare and Graph

```python
p = poisson.pmf(s,P)

plt.plot(s, px, '--r', label = 'Appro Prob')
plt.plot(s, p, 'b', label= 'Exact Prob')
plt.legend()
plt.show()
```
