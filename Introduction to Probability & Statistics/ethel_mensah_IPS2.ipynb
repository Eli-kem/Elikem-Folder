```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

### 1a. Simulating M n-samples

```python
def Normal(mean,std, N,M):
    return np.random.normal(mean,std, size=(N,M))
```

```python
mean = 5
std = 0.5
N = 100
M = 1000

U = Normal(mean,std, N,M)
U
```

### 1b.The mean for each column

```python
mean_U =np.mean(U, axis = 1)
mean_U
```

### 2a. The mean of the estimates

```python
Sum_mean = (np.mean(mean_U))
Sum_mean
```

### 2b. The mean square error 

```python
MSE = (Sum_mean- mean)**2    
print('The mean squared error =', MSE)
```

  Comment: The estimator is biased, since it does not equal the mean exactly


### 3. Construct a Histogram

```python
plt.hist(mean_U)
plt.show()
```

### 4. The variance for each column

```python
X2 =(np.mean(U**2, axis = 1))
Var_U = X2 - mean_U**2 
Var_U
```

### The variance of the estimates

```python
Sum_Var = np.mean(Var_U)
Sum_Var
```

### The Unbiased estimator

```python
Var = (N/(N-1))*Sum_Var
Var
```

### 5.


```python
Chi_sq = (N-1)*(1/0.25)*Var_U

plt.hist(Chi_sq)
plt.show()
```
